```python
#Encoding of Categorical Values
from sklearn import preprocessing
def encoding(df):
  for col in df:
    if df[col].dtype =='object':
      label_encoder = preprocessing.LabelEncoder()
      df[col]= label_encoder.fit_transform(df[col])

encoding(df_train)

df_train.head(10)



```

```python
#Selecting Features
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

X=df_train.drop(['class'], axis=1)
Y=df_train['class']

muinfocl=mutual_info_classif(X,Y)

muinfocl


mutual_info=pd.Series(muinfocl)

mutual_info.index=X.columns

mutual_info.sort_values(ascending=False)

mutual_info.sort_values(ascending=False).plot.bar(figsize=(50,45))

sel_cols = SelectKBest(mutual_info_classif,k=35)

sel_cols.fit(X,Y)

features_selected= X.columns[sel_cols.get_support()]

X=X[features_selected]


print(X)


```

```python
#Split the dataset
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=35)

sc=StandardScaler()

X_train=sc.fit_transform(X_train)

X_test = sc.transform(X_test)

print(X_train)

print(X_test)

```

```python
#Model Building

from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
#from sklearn.neural_network import MLPCLassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

#Applying Models
#Decision Tree Classifier
dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_train,Y_train)
y_predict_dtc= dtc.predict(X_test)

print("Decision Tree Classifier Accuracy =", accuracy_score(Y_test,y_predict_dtc)*100)
print("Decision Tree Classifier Precision =", precision_score(Y_test,y_predict_dtc)*100)
print("Decision Tree Classifier Recall =", recall_score(Y_test,y_predict_dtc)*100)
print("Decision Tree Classifier f1_score =", f1_score(Y_test,y_predict_dtc)*100)

from sklearn import tree

tree.plot_tree(dtc)

conf_mat = confusion_matrix(y_true=Y_test,y_pred=y_predict_dtc)

print(conf_mat)
print(classification_report(Y_test,y_predict_dtc))
```

```python
#Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB,BernoulliNB

#Gaussian Naive Bayes Algorithm
gnb = GaussianNB()
gnb = gnb.fit(X_train,Y_train)
y_predict_gnb= gnb.predict(X_test)

print("Gaussian Naive Bayes Accuracy =", accuracy_score(Y_test,y_predict_gnb)*100)
print("Gaussian Naive Bayes Precision =", precision_score(Y_test,y_predict_gnb)*100)
print("Gaussian Naive Bayes Recall =", recall_score(Y_test,y_predict_gnb)*100)
print("Gaussian Naive Bayes f1_score =", f1_score(Y_test,y_predict_gnb)*100)
print("\n")
conf_mat = confusion_matrix(y_true=Y_test,y_pred=y_predict_gnb)

print(conf_mat)

print(classification_report(Y_test,y_predict_gnb))

#Bernoulis Naive Bayes Algorithm
bnb = BernoulliNB()
bnb = bnb.fit(X_train,Y_train)
y_predict_bnb= bnb.predict(X_test)
print("Bernoullis Naive Bayes Accuracy =", accuracy_score(Y_test,y_predict_bnb)*100)
print("Bernoullis Naive Bayes Precision =", precision_score(Y_test,y_predict_bnb)*100)
print("Bernoullis Naive Bayes Recall =", recall_score(Y_test,y_predict_bnb)*100)
print("Bernoullis Naive Bayes f1_score =", f1_score(Y_test,y_predict_bnb)*100)
print("\n")

conf_mat = confusion_matrix(y_true=Y_test,y_pred=y_predict_bnb)

print(conf_mat)
print(classification_report(Y_test,y_predict_bnb))

```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr = lr.fit(X_train,Y_train)
y_predict_lr= lr.predict(X_test)

print("Logistic Regression Accuracy: -", accuracy_score(Y_test,y_predict_lr)*100)
print("Logistic Regression Precision: -", precision_score(Y_test,y_predict_lr)*100)
print("Logistic Regression Recall: -", recall_score(Y_test,y_predict_lr)*100)
print("Logistic Regression F1 Score: -", f1_score(Y_test,y_predict_lr)*100)

conf_mat = confusion_matrix(y_true=Y_test,y_pred=y_predict_lr)

print(conf_mat)

print(classification_report(Y_test,y_predict_lr))

```
